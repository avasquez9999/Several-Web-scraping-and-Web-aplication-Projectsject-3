```python
import requests
from bs4 import BeautifulSoup

url = 'https://catalog.fullerton.edu/content.php?catoid=61&navoid=7316'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

faculty_data = []

divs = soup.findAll("table", {"class": "table_default"})

for div in divs:
  # Extract information (assuming specific structure)
  name_tags = div.find_all('strong')
  position_tag = div.find('em')

  # Check if elements exist before accessing
  if name_tags and position_tag:
    name = name_tags[0].text.strip()  # Assuming first strong tag for name
    position = position_tag.text.strip()  # Assuming em tag holds position

    # Store data in a dictionary (or list)
    faculty_data.append({
      "Name": name,
      "Position": position
    })
  else:
    print("Error: Missing elements in current entry.")

# Print or use the faculty_data (list of dictionaries) for further processing
print(faculty_data)
```
